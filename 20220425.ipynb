```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa, librosa.display
```

```python
import os
url = "https://raw.githubusercontent.com/hsnam95/class2022Spring/main/aeiou.wav"
os.system("curl " + url + " > aeiou.wav")

s, sr = librosa.load('aeiou.wav')
```

```python
s = librosa.util.normalize(s)
librosa.display.waveplot(s, sr)
ipd.Audio(s[7000:12000], rate=sr)
```

```python
from scipy.signal import lfilter
sig = s[7000:12000]
sig = lfilter(np.array([1]), np.array([1]), sig, axis=0)
librosa.display.waveplot(sig, sr)
ipd.Audio(sig, rate=sr)
```

auto correlation(제곱평균제곱근)

; quadratic mean(이차평균)이라고도 하며, 변화하는 값의 크기에 대한 통계적 척도이다. 

-구하는 방법) 원래의 값의 제곱을 평균낸 것의 제곱근


-주기함수의 경우 전체 시간에 대한 제곱평균제곱근은 한 주기의 제곱평균제곱근과 같음


```python
rms = librosa.feature.rms(s)
plt.plot(rms[0])
```

auto correlation(자기상관)

; 시간 또는 공간적으로 연속된 일련의 관측치들간에 존재하는 상관관계 (현재의 상태가 과거와 미래의 상태에 밀접한 연관을 지니는 경우 자기상관이 있다고 함)

-일반적으로 패턴이 반복되는 일련의 데이터에서 발생함

-자기상관 함수의 값은 -1에서 1의 범위 사이에서 나타남(0에 가까울수록 선형적 관계가 없음)




```python
F0, voiced_flag, voiced_prob = librosa.pyin(s, 60, 200)
plt.plot(F0, '.')
```
