```python
def test(siDo):
  URL = ('http://apis.data.go.kr/B552061/AccidentDeath/'
        'getRestTrafficAccidentDeath?'
        '&ServiceKey=jXFdT16d5oT780rXsQYMwH%2BpfukLT5OV6fH6ELcLJv3uwOyNhfKRgWxrJCAize6wLtCEG9TOaDh9wNTh08oYiw%3D%3D'
        '&searchYear=2019'
        '&siDo='+siDo+
        '&guGun=1116'
        '&numOfRows=20'
        '&pageNo=1'
        '&type=json')
  
  response = requests.get(URL)
  r = response.json()
  items = r['items']['item']
            
  acc_ty_cd = [item['acc_ty_cd'] for item in items]
  acc_year = [item['acc_year'] for item in items]
  dth_dnv_cnt = [item['dth_dnv_cnt'] for item in items]
  occrrnc_lc_sido_cd = [item['occrrnc_lc_sido_cd'] for item in items]

  acc_ty_cd = pd.Series(acc_ty_cd)
  acc_year = pd.Series(acc_year)
  dth_dnv_cnt = pd.Series(dth_dnv_cnt)
  occrrnc_lc_sido_cd = pd.Series(occrrnc_lc_sido_cd)

  df = pd.DataFrame({ 'acc_ty_cd': acc_ty_cd, 'acc_year': acc_year, 'dth_dnv_cnt': dth_dnv_cnt, 'occrrnc_lc_sido_cd': occrrnc_lc_sido_cd})
  return df 
```

```python
iface = gr.Interface(test, gr.inputs.Dropdown(['01','25','25','22','02','25','05','04','05','32','23','21','02']), outputs=['dataframe'])
iface.launch(debug=True)
```
